### Questão 1

```python
# Filtrando divorciados 
divorciados = df.query("marital == 'divorced'")
```

```python
divorciados
```

```python
# Tabela de frequência cruzada de distribuição dos jobs de divorciados
tab_freq = pd.crosstab(divorciados['job'], divorciados['marital'])
tab_freq
```

```python
# Total de divorciados
total_divorciados = tab_freq.sum().values[0]
```

```python
# Extraindo clientes empreendedores e estudantes que são divorciados
empreendedores = tab_freq.loc['entrepreneur'].values[0] if 'entrepreneur' in tab_freq.index else 0
estudantes = tab_freq.loc['student'].values[0] if 'student' in tab_freq.index else 0
```

```python
# Percentuais de clientes empreendedores e divorciados / estudantes e divorciados
perc_empreendedores_divorciados = (empreendedores / total_divorciados) * 100
perc_estudantes_divorciados = (estudantes / total_divorciados) * 100
```

```python
print(f"Percentual de divorciados que são empreendedores: {perc_empreendedores_divorciados:.2f}%")
print(f"Percentual de divorciados que são estudantes: {perc_estudantes_divorciados:.2f}%")
```

## Questão 2

O seu chefe decidiu criar duas categorias, Premium e Gold, para focar nos clientes mais promissores. Para a categoria Premium, entrarão os que possuem saldo (balance) nos top 10%. Ou seja, o grupo acima do percentil 90. Já o grupo Gold, ficarão os que possuem saldo nos top 20%, mas abaixo do grupo Premium. Ou seja, do p80 até o último do p90. Verifique quantos clientes possuem no grupo Premium e no Gold.

```python
# Definindo percentil 80 e 90
percentil_80 = df['balance'].quantile(0.80)
percentil_90 = df['balance'].quantile(0.90)
```

```python
# Categorizando os clientes Gold e Premium
df['category'] = pd.qcut(df['balance'], q=[0.0, 0.8, 0.9, 1.0], labels=['Regular', 'Gold', 'Premium'])
```

```python
# Número de clientes Gold e Premium
gold_clientes = df[df['category'] == 'Gold'].shape[0]
premium_clientes = df[df['category'] == 'Premium'].shape[0]

print("Clientes Gold:", gold_clientes)
print("Clientes Premium:", premium_clientes)
```

## Questão 3

Seu chefe também quer saber se existe diferença nos percentuais vistos no exercício (1) para o grupo Premium. Verifique como a tabela cruzada da população se diferencia do grupo Premium. Existe diferença? Qual sua hipótese para que exista essa diferença?

```python
# Filtrando clientes premium
premium = df[df['category'] == 'Premium']
```

```python
# Filtrando os clientes Premium divorciados
premium_divorciados = premium.query("marital == 'divorced'")
```

```python
# Frequência cruzada divorciados premium
tab_freq_divorciados_premium = pd.crosstab(premium_divorciados['job'], premium_divorciados['marital'])

total_divorciados_premium = tab_freq_divorciados_premium.sum().values[0]
```

```python
empreendedores_divorciados_premium = tab_freq_divorciados_premium.loc['entrepreneur'].values[0] if 'entrepreneur' in tab_freq_divorciados_premium.index else 0

estudantes_divorciados_premium = tab_freq_divorciados_premium.loc['student'].values[0] if 'student' in tab_freq_divorciados_premium.index else 0
```

```python
# Definindo o percentual de empreendedores e estudantes do grupo Premium
perc_empreendedores_divorciados_premium = (empreendedores_divorciados_premium / total_divorciados_premium) * 100
perc_estudantes_divorciados_premium = (estudantes_divorciados_premium / total_divorciados_premium) * 100

print(f"Percentual de divorciados que são empreendedores no grupo Premium: {perc_empreendedores_divorciados_premium:.2f}%")
print(f"Percentual de divorciados que são estudantes no grupo Premium: {perc_estudantes_divorciados_premium:.2f}%")
```

```python
# Comparação
print(f"Empreendedores - População geral: {perc_empreendedores_divorciados:.2f}%, Clientes Premium: {perc_empreendedores_divorciados_premium:.2f}%")
print(f"Estudantes - População geral: {perc_estudantes_divorciados:.2f}%, Clientes Premium: {perc_estudantes_divorciados_premium:.2f}%")
```

## Questão 4

Para a apresentação final, você decide criar uma tabela com os 10 clientes que possuem maior saldo (balance). Como a tabela é grande demais, você resolve apresentar apenas as informações relevantes: age, job, marital, education, default, loan, duration. Além disso, você decide destacar clientes que foram mais e menos atenciosos, pois são os que parecem ter maior e menor potencial, respectivamente. Para isso, destaque na tabela, de amarelo, o cliente de maior duration. Destaque, de vermelho, o que tem menor duration.

```python
# Clientes com maior saldo
top_10_clientes = df.nlargest(10, 'balance')
```

```python
# Colunas relevantes
colunas_relevantes = ['age', 'job', 'marital', 'education', 'default', 'loan', 'duration']
top_10_clientes = top_10_clientes[colunas_relevantes]
```

```python
# Clientes mais e menos atenciosos (maior e menor potencial)
clientes_mais_atenc = top_10_clientes['duration'].idxmax()
clientes_menos_atenc = top_10_clientes['duration'].idxmin()
```

```python
# Função para destacar os clientes
def highlight_max_min(s):
    color = [''] * len(s)
    if s.name == clientes_mais_atenc:
        color = ['background-color: yellow'] * len(s)
    elif s.name == clientes_menos_atenc:
        color = ['background-color: red'] * len(s)
    return color
```

```python
# Aplicando a formatação condicional e imprimindo a tabela (amarelo = maior duration / vermelho = menor duration)
styled_df = top_10_clientes.style.apply(highlight_max_min, axis=1)
styled_df
```

## Questão 5

Na sua opinião e de parte da diretoria, há 3 fatores que determinam se a pessoa vai aceitar a proposta do seu banco (y): o saldo (balance), se ela já deixou de pagar um empréstimo (default) e se ela tem um empréstimo pessoal (loan). Para avaliar essas 3 características junto com o aceite (y), você precisará criar uma tabela dinâmica, cujo valor observado seja a média do saldo. Avalie se existe diferença no saldo de quem aceita a proposta e tem empréstimo vs quem não aceita e não tem empréstimo, e o mesmo para o default. Em outras palavras, crie uma tabela dinâmica cruzando loan e default, mas também segmentando por aceite ou não da proposta. O saldo de quem aceita e tem empréstimo é maior ou menor do que o de quem não aceita? Por qual motivo você acha que isso acontece? Crie a sua hipótese para justificar o resultado achado!

```python
# Tabela dinâmica com média de saldo
pivot_table = pd.pivot_table(df, 
                             values='balance', 
                             index=['loan', 'default'], 
                             columns='y', 
                             aggfunc=np.mean)
```

```python
pivot_table
```

#### Conclusão:
A média do saldo de quem aceita a proposta e tem empréstimo é MENOR do que a média do saldo de quem não aceita a proposta e não tem empréstimo.
Acredito que isso aconteça pois quem não tem empréstimo ativo não compromete seu saldo para pagar parcelas, fazendo com que seu dinheiro se mantenha na conta e a média seja mais alta.
